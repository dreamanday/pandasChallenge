### Top Spenders

```python
top_nerds = data.groupby('SN').count()
top_nerds = top_nerds.sort_values(by=['Purchase ID'],ascending=False)
top_nerds = top_nerds['Purchase ID']
top_nerds = top_nerds.head(5)
top_nerds = pd.DataFrame(top_nerds)
top_nerds.reset_index(inplace = True)
top_nerds
```

```python
nerds=[]
count_nerds=[]
avg_nerds=[]
sum_nerds=[]

x=0
y=0
z=0

for index, row in top_nerds['SN'].iteritems():
    for a,b,c,d,e,f,g in data.itertuples(index=False):
        if row == b:
            y=y+1
            x=x+g
            z=x/y
    nerds.append(row)
    count_nerds.append(y)
    avg_nerds.append(z)
    sum_nerds.append(x)
    x=0
    y=0
    z=0

        
total_nerds = pd.DataFrame({
    
    'Nerd': nerds,
    'Purchase Count': count_nerds,
    'Average Purchase Price': avg_nerds,
    'Total Purchase Value': sum_nerds
    
})    

total_nerds
```
